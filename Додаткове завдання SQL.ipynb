**ЗАДАЧА8.**  **Знайдіть співробітників з декількома посадами**
    
    Деякі співробітники можуть змінювати свої посади протягом роботи в компанії. Визначте всіх співробітників, які мали більше однієї посади за час роботи в компанії. Відобразіть їх `emp_no`, `first_name`, `last_name` та список всіх посад, на яких вони працювали.

```python
sql = """SELECT emp.emp_no, emp.first_name, emp.last_name, GROUP_CONCAT(tit.title) AS titles
FROM employees AS emp
JOIN titles AS tit ON emp.emp_no = tit.emp_no
GROUP BY emp.emp_no, emp.first_name, emp.last_name
HAVING COUNT(DISTINCT tit.title) > 1;"""
```

```python
df = pd.read_sql(sql, con)
```

```python
df
```

**ЗАДАЧА9.** 1. **Переводи між відділами**
    
    Співробітники іноді переходять між відділами. Визначте всіх співробітників, які працювали в більш ніж одному відділі. Відобразіть їх `emp_no`, `first_name`, `last_name` та назви всіх відділів, в яких вони працювали.
    

```python
sql = """SELECT emp.emp_no, emp.first_name, emp.last_name, GROUP_CONCAT(DISTINCT dept.dept_name) AS departments
FROM employees AS emp
JOIN dept_emp AS dept_emp ON emp.emp_no = dept_emp.emp_no
JOIN departments AS dept ON dept_emp.dept_no = dept.dept_no
GROUP BY emp.emp_no, emp.first_name, emp.last_name
HAVING COUNT(DISTINCT dept_emp.dept_no) > 1;"""
```

```python
df = pd.read_sql(sql, con)
```

```python
df
```

**Window Functions** ЗАДАЧА 1 Використання ROW_NUMBER
Для перших 10 співробітників (за emp_no) виведіть їх emp_no, first_name, last_name, а також порядковий номер співробітника в таблиці.

```python
sql = """SELECT emp_no, first_name, last_name,
       ROW_NUMBER() OVER (ORDER BY emp_no) as row_numb
FROM employees
LIMIT 10;"""
```

```python
sql = """SELECT e.emp_no, d.dept_name, s.salary,
RANK() OVER (PARTITION BY de.dept_no ORDER BY s.salary DESC) as salary_rank
FROM employees e
JOIN salaries s ON e.emp_no = s.emp_no
JOIN dept_emp de ON e.emp_no = de.emp_no
JOIN departments d ON de.dept_no = d.dept_no
WHERE s.to_date = '9999-01-01'
LIMIT 10;"""
```
